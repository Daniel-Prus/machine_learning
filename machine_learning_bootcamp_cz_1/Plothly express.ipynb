```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons
import plotly.express as px


np.random.seed(42)
raw_data = make_moons(n_samples=700, noise=0.25, random_state=42)
data = raw_data[0]
target = raw_data[1]
df = pd.DataFrame(data=np.c_[data, target], columns=['x1', 'x2', 'target'])
```

```python
# color scales to choose

fig = px.colors.sequential.swatches_continuous()
fig
```

```python
# scatter

px.scatter(df,x='x1', y='x2', color='target', title='Classification x1, x2', 
           color_continuous_scale=px.colors.sequential.Viridis, height = 500, width=800)

# opacity=0.6 - odpowiednik alpha
```

```python
# eby podzia 'targetu' by bez tej kolorowej osi, target musi by typem 'str' (doda kolumn z nazwami 0,1)
fig = px.scatter_matrix(df,dimensions=list(df.columns)[:5], color="target")

fig.update_layout(
    font=dict(
        #family="Courier New, monospace",
        size=18,
        color="RebeccaPurple"))
        
fig.show()
```
